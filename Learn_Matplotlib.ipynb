```python
data_set_size = 15
low_mu, low_sigma = 50,4.3
low_data_set = low_mu + low_sigma * np.random.random(data_set_size)
high_mu, high_sigma = 57,5.2
high_data_set = high_mu + high_sigma * np.random.random(data_set_size)

days = list(range(1,data_set_size + 1))
```

```python
plt.plot(days,low_data_set)
plt.show()
```

```python
plt.plot(days, low_data_set,days, high_data_set)
plt.figure()
```

```python
number_of_data = 1000
plt1 = plt.figure()
subplot_1 = plt1.add_subplot(1,1,1)
subplot_1.plot(np.random.rand(number_of_data).cumsum())

number_of_ticks = 5
tics = np.arange(0,number_of_data,number_of_data // number_of_ticks)
subplot_1.set_xticks(tics)

labels = subplot_1.set_xticklabels(['one','two','three','four','five'], rotation = 45, fontsize = 'small')

subplot_1.set_title("My first ticked plot")
subplot_1.set_xlabel("Groups")

subplot_1.grid(True)

gridlines = subplot_1.get_xgridlines() + subplot_1.get_ygridlines()
for line in gridlines:
    line.set_linestyle(':')
plt.show()
```

```python
fig = plt.figure()

for i,label in enumerate(('A','B','C','D')):
    ax = fig.add_subplot(2,2,i+1)
    ax.text(0.05,0.95, label, transform = ax.transAxes,
           fontsize = 16, fontweight = 'bold',va = 'top')

plt.show()
```

```python
labels = 'Frogs','Hogs','Dogs','Logs'
sizes = [15,30,45,10]
colors = ['yellowgreen','gold','lightskyblue','lightcoral']
explode = (0,0.1,0,0)

plt.pie(x=sizes,explode= explode, labels = labels, colors = colors,
       autopct = '%1.1f%%',shadow = True, startangle = 90)
plt.axis('equal')

plt.show()
```
